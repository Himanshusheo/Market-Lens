```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from datetime import datetime, timedelta
import calendar

# Load the data
orders_data = pd.read_csv('Monthly_Master (1).csv')
monthly_data = pd.read_csv('Final_monthly.csv')

# Convert dates to datetime format
if 'order_date_only' in orders_data.columns:
    orders_data['order_date_only'] = pd.to_datetime(orders_data['order_date_only'])
monthly_data['Month'] = pd.to_datetime(monthly_data['Month'])

# Analyze GMV seasonality to determine number of sales events per month
# First, aggregate orders by month to see monthly GMV patterns
orders_data['month'] = orders_data['order_date_only'].dt.month
orders_data['year'] = orders_data['order_date_only'].dt.year

# Calculate total GMV by month
monthly_gmv = orders_data.groupby(['year', 'month'])['gmv'].sum().reset_index()

# Create a function to normalize GMV and determine suggested number of sale events
def determine_sale_events_count(monthly_gmv):
    """
    Determine the number of sale events per month based on GMV seasonality
    """
    # Get the average GMV per month across all available years
    avg_monthly_gmv = monthly_gmv.groupby('month')['gmv'].mean().reset_index()
    
    # Normalize GMV to a scale of 0-3 (for number of events)
    min_gmv = avg_monthly_gmv['gmv'].min()
    max_gmv = avg_monthly_gmv['gmv'].max()
    
    avg_monthly_gmv['gmv_norm'] = (avg_monthly_gmv['gmv'] - min_gmv) / (max_gmv - min_gmv) * 3
    
    # Round to nearest integer with minimum of 1 for months with very low GMV
    avg_monthly_gmv['suggested_events'] = np.round(avg_monthly_gmv['gmv_norm']).astype(int)
    avg_monthly_gmv['suggested_events'] = avg_monthly_gmv['suggested_events'].clip(lower=1)
    
    # Convert to dictionary for easy lookup - ensure keys are integers
    events_dict = dict(zip(avg_monthly_gmv['month'].astype(int), avg_monthly_gmv['suggested_events']))
    
    return events_dict, avg_monthly_gmv

# Get suggested number of sale events per month
events_dict, avg_monthly_gmv = determine_sale_events_count(monthly_gmv)

# Display seasonality patterns and suggested events
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=avg_monthly_gmv, x='month', y='gmv')
ax2 = ax.twinx()
sns.lineplot(data=avg_monthly_gmv, x='month', y='suggested_events', color='red', marker='o', ax=ax2)
ax.set_xlabel('Month')
ax.set_ylabel('Average GMV')
ax2.set_ylabel('Suggested # of Sale Events')
plt.title('GMV Seasonality and Suggested Number of Sale Events')
months = list(calendar.month_abbr)[1:]
plt.xticks(range(12), months)
plt.tight_layout()
plt.show()

print("Suggested number of sale events per month based on GMV seasonality:")
for month, count in sorted(events_dict.items()):
    # Convert month to integer before accessing calendar.month_name
    month_name = calendar.month_name[int(month)]
    print(f"{month_name}: {count} events")

# Define the number of sale events for 2024-2025 based on our analysis
sales_events_per_month = {}

# For 2024 (June-December)
for month in range(6, 13):
    sales_events_per_month[(2024, month)] = events_dict.get(month, 1)

# For 2025 (January-June)
for month in range(1, 7):
    sales_events_per_month[(2025, month)] = events_dict.get(month, 1)

print("\nPlanned sales events for 2024-2025:")
for (year, month), count in sorted(sales_events_per_month.items()):
    month_name = calendar.month_name[month]
    print(f"{month_name} {year}: {count} events")

# Function to score potential sale dates
def score_sale_date(date, historical_data, holidays_df, monthly_model_data):
    """
    Score a potential sale date based on multiple factors
    """
    date_obj = pd.Timestamp(date)
    month = date_obj.month
    day_of_week = date_obj.dayofweek  # 0=Monday, 6=Sunday
    day_of_month = date_obj.day
    
    score = 0
    
    # Day of week preference based on historical data
    dow_scores = {
        0: 15.91,  # Monday
        1: 20.45,  # Tuesday
        2: 18.18,  # Wednesday
        3: 13.64,  # Thursday
        4: 9.09,   # Friday
        5: 11.36,  # Saturday
        6: 11.36   # Sunday
    }
    score += dow_scores.get(day_of_week, 5)
    
    # Holiday proximity score
    is_holiday = date_obj.date() in holidays_df['date'].dt.date.values
    if is_holiday:
        score += 5
    
    # Check for holidays within 3 days
    for delta in range(-3, 4):
        check_date = date_obj + pd.Timedelta(days=delta)
        if check_date.date() in holidays_df['date'].dt.date.values:
            score += max(0, 5 - abs(delta))  # Higher score for closer holidays
    
    # Monthly seasonality score based on historical data
    month_scores = {
        1: 13.64,   # January
        2: 13.64,   # February
        3: 6.82,    # March
        5: 6.82,    # May
        7: 4.55,    # July
        8: 13.64,   # August
        10: 6.82,   # October
        11: 18.18,  # November
        12: 15.91   # December
    }
    score += month_scores.get(month, 5)
    
    # Historical sales performance for this month
    month_data = monthly_model_data[monthly_model_data['Month'].dt.month == month]
    if not month_data.empty:
        avg_units = month_data['units'].mean()
        # Normalize to a 0-15 scale
        sales_score = min(15, avg_units / 1000)
        score += sales_score
    
    # Avoid scheduling sales too close to previous sales
    for _, row in historical_data.iterrows():
        hist_date = row['date'].date()
        days_diff = abs((date_obj.date() - hist_date).days)
        if days_diff < 14:  # Penalize dates within 2 weeks of historical sales
            score -= max(0, 10 - days_diff/2)
    
    return score

# Generate GMV-based seasonal weights for monthly marketing budgets
def generate_budget_factors(monthly_gmv):
    """
    Generate budget factors based on monthly GMV trends
    """
    # Get average monthly GMV
    avg_gmv = monthly_gmv.groupby('month')['gmv'].mean().reset_index()
    
    # Calculate relative budgets (as multipliers)
    total_avg = avg_gmv['gmv'].mean()
    avg_gmv['budget_factor'] = avg_gmv['gmv'] / total_avg
    
    # Scale to have average of 1.0
    scaled_factors = avg_gmv['budget_factor'] / avg_gmv['budget_factor'].mean()
    
    # Create dictionary with integer month keys
    return dict(zip(avg_gmv['month'].astype(int), scaled_factors))

# Generate monthly budget factors
monthly_budget_factors = generate_budget_factors(monthly_gmv)

print("\nMonthly Budget Factors based on GMV Seasonality:")
for month, budget in sorted(monthly_budget_factors.items()):
    # Fix the type error by ensuring month is an integer
    month_name = calendar.month_name[int(month)]
    print(f"{month_name}: {budget:.2f}x")

# Create total budget dictionary for the prediction period
# Define base monthly budget (adjust as needed)
base_monthly_budget = 100000  # example value

total_monthly_budgets = {}
for year in [2024, 2025]:
    start_month = 6 if year == 2024 else 1
    end_month = 13 if year == 2024 else 7
    
    for month in range(start_month, end_month):
        # Apply seasonal scaling factor
        factor = monthly_budget_factors.get(month, 1.0)
        total_monthly_budgets[(year, month)] = base_monthly_budget * factor

print("\nTotal Monthly Budgets for 2024-2025:")
for (year, month), budget in sorted(total_monthly_budgets.items()):
    month_name = calendar.month_name[month]
    print(f"{month_name} {year}: ${budget:.2f}")

# Define marketing channels
marketing_channels = ['TV', 'Digital', 'Sponsorship', 'Content Marketing', 
                     'Online marketing', " Affiliates", 'SEM', 'Radio', 'Other']

# Analyze channel effectiveness using Random Forest model
X = monthly_data[marketing_channels]
y = monthly_data['units']

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# Get feature importance
channel_importance = pd.DataFrame({
    'Channel': marketing_channels,
    'Importance': model.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nMarketing Channel Effectiveness:")
for i, row in channel_importance.iterrows():
    print(f"{row['Channel']}: {row['Importance']:.4f}")

# Function to optimize channel allocation for a given month
def optimize_monthly_allocation(channels, importance, total_budget):
    """
    Optimize budget allocation across channels based on importance
    """
    # Ensure positive importance values
    importance_values = np.maximum(importance, 0.001)
    
    # Calculate allocation based on relative importance
    total_importance = sum(importance_values)
    allocation = {ch: (imp/total_importance) * total_budget 
                 for ch, imp in zip(channels, importance_values)}
    
    return allocation

# Optimize budget allocation for each month
optimal_allocations = {}
for (year, month), budget in total_monthly_budgets.items():
    # Get channel importance values
    importance_values = channel_importance['Importance'].values
    
    # Adjust importance for seasonal factors
    # For example, digital might be more important during holiday seasons
    seasonal_adjustments = {
        11: {'Digital': 1.3, 'SEM': 1.2},  # November
        12: {'Digital': 1.4, 'Content Marketing': 1.3},  # December
        1: {'TV': 1.2, 'Radio': 1.1},  # January
        8: {'Digital': 1.2, 'Content Marketing': 1.2}  # August
    }
    
    # Apply seasonal adjustments to importance values
    adjusted_importance = importance_values.copy()
    if month in seasonal_adjustments:
        for i, channel in enumerate(marketing_channels):
            if channel in seasonal_adjustments[month]:
                adjusted_importance[i] *= seasonal_adjustments[month][channel]
    
    # Get optimized allocation
    allocation = optimize_monthly_allocation(
        marketing_channels, adjusted_importance, budget)
    
    optimal_allocations[(year, month)] = allocation

# Create a calendar of future dates to score
future_start = '2024-06-01'
future_end = '2025-06-30'
future_calendar = pd.DataFrame({
    'date': pd.date_range(start=future_start, end=future_end)
})
future_calendar['month'] = future_calendar['date'].dt.month
future_calendar['year'] = future_calendar['date'].dt.year
future_calendar['day_of_week'] = future_calendar['date'].dt.dayofweek
future_calendar['day_of_month'] = future_calendar['date'].dt.day
future_calendar['is_holiday'] = future_calendar['date'].dt.date.isin(holiday_filtered_df['date'].dt.date)

# Score each date
future_calendar['sale_score'] = future_calendar.apply(
    lambda row: score_sale_date(row['date'], historical_sales_df, holiday_filtered_df, monthly_data), 
    axis=1
)

# Define typical duration for sales in each month (based on historical data)
sale_duration_by_month = {
    1: 3,   # January
    2: 2,   # February
    3: 3,   # March
    5: 3,   # May
    6: 3,   # June
    7: 2,   # July
    8: 3,   # August
    9: 3,   # September
    10: 3,  # October
    11: 8,  # November
    12: 10  # December
}

# Select optimal sale dates
optimal_sales = []

for year in [2024, 2025]:
    # Define month range based on our prediction period
    if year == 2024:
        months = range(6, 13)  # June to December 2024
    else:
        months = range(1, 7)   # January to June 2025
        
    for month in months:
        # Skip months with no planned sales
        events_count = sales_events_per_month.get((year, month), 0)
        if events_count == 0:
            continue
            
        # Filter calendar for this month
        month_calendar = future_calendar[(future_calendar['year'] == year) & 
                                       (future_calendar['month'] == month)].copy()
        
        for event in range(events_count):
            if len(month_calendar) == 0:
                continue
                
            # Find best starting date
            best_start = month_calendar.sort_values('sale_score', ascending=False).iloc[0]['date']
            
            # Get appropriate duration
            duration = sale_duration_by_month.get(month, 3)
            
            # Calculate end date
            best_end = best_start + pd.Timedelta(days=duration-1)
            
            # Add to optimal sales list
            optimal_sales.append({
                'start_date': best_start,
                'end_date': best_end,
                'year': year,
                'month': month,
                'duration': duration,
                'score': month_calendar[month_calendar['date'] == best_start]['sale_score'].values[0]
            })
            
            # Remove selected dates and buffer period from consideration
            sale_period = pd.date_range(start=best_start-pd.Timedelta(days=7), 
                                     end=best_end+pd.Timedelta(days=7))
            month_calendar = month_calendar[~month_calendar['date'].isin(sale_period)]

# Create DataFrame for optimal sales
optimal_sales_df = pd.DataFrame(optimal_sales)
optimal_sales_df = optimal_sales_df.sort_values(['year', 'month'])

# Format dates for display
optimal_sales_df['start_formatted'] = optimal_sales_df['start_date'].dt.strftime('%b %d, %Y')
optimal_sales_df['end_formatted'] = optimal_sales_df['end_date'].dt.strftime('%b %d, %Y')

print("\nOptimal Sale Dates:")
print(optimal_sales_df[['year', 'month', 'start_formatted', 'end_formatted', 'duration', 'score']])

# Integrate marketing budget with sale dates
sales_calendar_data = []

for _, row in optimal_sales_df.iterrows():
    year = row['year']
    month = row['month']
    
    # Get the base allocation for this month
    base_allocation = optimal_allocations.get((year, month), {})
    
    # Increase budget for sale period (typically by 30%)
    sale_allocation = {channel: amount * 1.3 for channel, amount in base_allocation.items()}
    
    # Add to calendar
    sales_calendar_data.append({
        'year': year,
        'month': month,
        'start_date': row['start_date'],
        'end_date': row['end_date'],
        'duration': row['duration'],
        'score': row['score'],
        'budget_allocation': sale_allocation,
        'total_budget': sum(sale_allocation.values())
    })

# Create the calendar DataFrame
sales_calendar_df = pd.DataFrame(sales_calendar_data)

# Format the calendar for display
print("\n INTEGRATED SALES & MARKETING CALENDAR ")
for _, row in sales_calendar_df.sort_values(['year', 'month']).iterrows():
    start_str = row['start_date'].strftime('%b %d, %Y')
    end_str = row['end_date'].strftime('%b %d, %Y')
    
    print(f"\n{row['year']}-{row['month']:02d} SALE EVENT: {start_str} to {end_str} ({row['duration']} days)")
    print(f"Total Budget: ${row['total_budget']:.2f}")
    
    # Display top channels by budget
    allocation = row['budget_allocation']
    top_channels = sorted(allocation.items(), key=lambda x: x[1], reverse=True)[:3]
    
    print("Top Marketing Channels:")
    for channel, amount in top_channels:
        percentage = amount / row['total_budget'] * 100
        print(f"  {channel}: ${amount:.2f} ({percentage:.1f}%)")

# Create visualization of the calendar
plt.figure(figsize=(15, 10))

# Plot 1: Timeline of Sale Events
plt.subplot(2, 1, 1)
for i, row in sales_calendar_df.iterrows():
    start = row['start_date']
    duration = (row['end_date'] - row['start_date']).days + 1
    plt.barh(0, duration, left=start, height=0.3, color='blue', alpha=0.7)
    plt.text(start, 0.1, f"{row['month']}/{row['year']}", fontsize=8, color='black')

# Add holidays as vertical lines
for _, row in holiday_filtered_df.iterrows():
    holiday_date = row['date']
    if pd.Timestamp('2024-06-01') <= holiday_date <= pd.Timestamp('2025-06-30'):
        plt.axvline(x=holiday_date, color='red', linestyle='--', alpha=0.3)
        plt.text(holiday_date, 0.4, row['holiday_name'], rotation=90, fontsize=8)

plt.title('Sale Events Timeline (2024-2025)')
plt.yticks([])
plt.grid(True, axis='x')

# Plot 2: Monthly Marketing Budget by Channel
plt.subplot(2, 1, 2)

# Prepare data for visualization
months = [(2024, m) for m in range(6, 13)] + [(2025, m) for m in range(1, 7)]
month_labels = [f"{y}-{m:02d}" for y, m in months]
channel_data = {channel: [] for channel in marketing_channels}

for year, month in months:
    allocation = optimal_allocations.get((year, month), {})
    for channel in marketing_channels:
        channel_data[channel].append(allocation.get(channel, 0))

# Plot stacked bars
bottom = np.zeros(len(months))
for channel in marketing_channels:
    plt.bar(month_labels, channel_data[channel], bottom=bottom, label=channel)
    bottom += np.array(channel_data[channel])

plt.title('Monthly Marketing Budget by Channel')
plt.xlabel('Month')
plt.ylabel('Budget ($)')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()

# Export the calendar to CSV
export_data = []
for _, row in sales_calendar_df.iterrows():
    # Create a basic row with sale info
    export_row = {
        'Year': row['year'],
        'Month': row['month'],
        'Start Date': row['start_date'].strftime('%Y-%m-%d'),
        'End Date': row['end_date'].strftime('%Y-%m-%d'),
        'Duration': row['duration'],
        'Total Budget': row['total_budget']
    }
    
    # Add channel budgets
    for channel in marketing_channels:
        export_row[f'{channel} Budget'] = row['budget_allocation'].get(channel, 0)
    
    export_data.append(export_row)

export_df = pd.DataFrame(export_data)
export_df.to_csv('sale_and_marketing_plan_2024_2025.csv', index=False)
print("\nExported plan to 'sale_and_marketing_plan_2024_2025.csv'")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import calendar

# Modified function to map projected revenue to sale duration with reduced days
def revenue_to_sale_duration(projected_revenue):
    """
    Convert projected revenue value to recommended sale duration in days
    New Scale: 
    < 4.0  No sale
    4.0-5.0  2 days
    5.0-6.0  3 days
    6.0+  4 days
    """
    if projected_revenue < 4.0:
        return 0  # No sale for low revenue months
    elif projected_revenue < 5.0:
        return 2
    elif projected_revenue < 6.0:
        return 3
    else:
        return 4

# Create a dataframe with the projected revenue data
projected_revenue = {
    (2023, 5): 5.92,
    (2023, 6): 4.87,
    (2023, 7): 3.43,
    (2023, 8): 1.91,
    (2023, 9): 6.29,
    (2023, 10): 7.34,
    (2023, 11): 5.18,
    (2023, 12): 6.48,
    (2024, 1): 5.83,
    (2024, 2): 5.07,
    (2024, 3): 6.36,
    (2024, 4): 5.36,
    (2024, 5): 5.92,
    (2024, 6): 4.87,
    (2024, 7): 3.43
}

# Create sale duration mapping based on projected revenue
sale_duration = {}
for period, revenue in projected_revenue.items():
    # Special handling for holiday season with slightly longer duration
    year, month = period
    base_duration = revenue_to_sale_duration(revenue)
    if month in [11, 12] and base_duration > 0:  # Holiday season
        sale_duration[period] = min(5, base_duration + 1)  # Add 1 day for holiday season, max 5 days
    else:
        sale_duration[period] = base_duration

# Create a dataframe for easier visualization
sale_planning_df = pd.DataFrame([
    {'Year': year, 
     'Month': month, 
     'Month_Name': calendar.month_name[month],
     'Projected_Revenue': projected_revenue[(year, month)],
     'Recommended_Sale_Duration': sale_duration[(year, month)]} 
    for year, month in sorted(projected_revenue.keys())
])

# Round to 2 decimal places for display
sale_planning_df['Projected_Revenue'] = sale_planning_df['Projected_Revenue'].round(2)

# Modified function to calculate number of sales events
def calculate_sale_events(df):
    df = df.copy()
    
    # Initialize Num_Sale_Events to 0
    df['Num_Sale_Events'] = 0
    
    # Group by year and rank months by projected revenue
    df['Revenue_Rank'] = df.groupby('Year')['Projected_Revenue'].rank(ascending=False)
    
    # Only assign sales to months with non-zero duration
    df.loc[df['Recommended_Sale_Duration'] > 0, 'Num_Sale_Events'] = 1
    
    # Top 2 revenue months per year get 2 events
    df.loc[(df['Revenue_Rank'] <= 2) & (df['Recommended_Sale_Duration'] > 0), 'Num_Sale_Events'] = 2
    
    # Holiday months (Nov & Dec) always get 2 events if they have non-zero duration
    df.loc[(df['Month'].isin([11, 12])) & (df['Recommended_Sale_Duration'] > 0), 'Num_Sale_Events'] = 2
    
    return df

# Calculate recommended number of sale events
sale_planning_df = calculate_sale_events(sale_planning_df)

# Calculate the suggested number of sale days per month
sale_planning_df['Total_Sale_Days'] = sale_planning_df['Recommended_Sale_Duration'] * sale_planning_df['Num_Sale_Events']

# Modified function to suggest optimal days for sales
def suggest_optimal_days(month, duration, num_events):
    """Suggest the optimal days for sales based on day-of-week preferences"""
    if duration == 0 or num_events == 0:
        return []
        
    # Define day of week preference (0=Monday, 6=Sunday)
    best_start_days = [1, 2]  # Tuesday, Wednesday
    
    suggestions = []
    
    for event in range(num_events):
        start_day = best_start_days[event % 2]
        
        # For second event in the month, shift to middle/end of month
        if event == 0:
            week_start = "second"  # Changed from first to second week
        else:
            week_start = "fourth"  # Changed from third to fourth week
            
        suggestions.append(f"{week_start} {calendar.day_name[start_day]} + {duration} days")
    
    return suggestions

# Add suggested optimal days
sale_planning_df['Suggested_Days'] = sale_planning_df.apply(
    lambda x: suggest_optimal_days(x['Month'], x['Recommended_Sale_Duration'], x['Num_Sale_Events']), 
    axis=1
)

# Create visualization
plt.figure(figsize=(12, 8))
ax = plt.subplot(111)

# Prepare data for visualization
sale_planning_pivot = sale_planning_df.copy()
sale_planning_pivot['Month_Year'] = sale_planning_pivot.apply(lambda x: f"{x['Month_Name'][:3]} {x['Year']}", axis=1)

# Create bars for each event
for event in range(1, sale_planning_df['Num_Sale_Events'].max()+1):
    event_durations = []
    for _, row in sale_planning_pivot.iterrows():
        if event <= row['Num_Sale_Events']:
            event_durations.append(row['Recommended_Sale_Duration'])
        else:
            event_durations.append(0)
    
    if event == 1:
        bottom = None
        label = "First Event"
    else:
        bottom = np.array(prev_heights)
        label = f"Event #{event}"
    
    bars = plt.bar(sale_planning_pivot['Month_Year'], event_durations, bottom=bottom, label=label)
    
    if event == 1:
        prev_heights = event_durations
    else:
        prev_heights = [a + b for a, b in zip(prev_heights, event_durations)]

# Customize the chart
plt.title('Recommended Sale Days by Month')
plt.xlabel('Month')
plt.ylabel('Total Sale Days')
plt.xticks(rotation=45)
plt.legend(title="Sale Events")
plt.tight_layout()

# Add sale duration details
for i, row in enumerate(sale_planning_pivot.itertuples()):
    if row.Total_Sale_Days > 0:
        plt.text(i, row.Total_Sale_Days + 0.5, f"{row.Total_Sale_Days} days", 
                ha='center', va='bottom', fontweight='bold')

# Modified calendar creation function
def create_sale_calendar(planning_df, start_year=2024, start_month=6, end_year=2025, end_month=6):
    """Create a calendar with specific sale dates based on the planning data"""
    sale_calendar = []
    
    for _, row in planning_df.iterrows():
        if row['Recommended_Sale_Duration'] == 0 or row['Num_Sale_Events'] == 0:
            continue
            
        year, month = row['Year'], row['Month']
        
        # Skip if outside of our planning period
        if (year < start_year or (year == start_year and month < start_month) or
            year > end_year or (year == end_year and month > end_month)):
            continue
            
        num_events = row['Num_Sale_Events']
        duration = row['Recommended_Sale_Duration']
        
        # Get the first day of the month
        first_day = datetime(year, month, 1)
        
        # Find the second Tuesday and fourth Wednesday
        days_until_tuesday = (1 - first_day.weekday()) % 7
        second_tuesday = first_day + timedelta(days=days_until_tuesday + 7)
        
        days_until_wednesday = (2 - first_day.weekday()) % 7
        fourth_wednesday = first_day + timedelta(days=days_until_wednesday + 21)
        
        # Event 1: Start on the second Tuesday
        start_date1 = second_tuesday
        end_date1 = start_date1 + timedelta(days=duration-1)
        
        sale_calendar.append({
            'Year': year,
            'Month': month,
            'Month_Name': calendar.month_name[month],
            'Event': 1,
            'Start_Date': start_date1,
            'End_Date': end_date1,
            'Duration': duration,
            'Projected_Revenue': row['Projected_Revenue']
        })
        
        # If there should be a second event
        if num_events >= 2:
            start_date2 = fourth_wednesday
            end_date2 = start_date2 + timedelta(days=duration-1)
            
            sale_calendar.append({
                'Year': year,
                'Month': month,
                'Month_Name': calendar.month_name[month],
                'Event': 2,
                'Start_Date': start_date2,
                'End_Date': end_date2,
                'Duration': duration,
                'Projected_Revenue': row['Projected_Revenue']
            })
    
    return pd.DataFrame(sale_calendar)

# Create the sales calendar with specific dates
sales_calendar = create_sale_calendar(sale_planning_df)

# Display the final sales calendar
print("\n#  FINAL SALES CALENDAR (2024-2025) ")
for month_year, group in sales_calendar.groupby(['Year', 'Month']):
    year, month = month_year
    month_name = calendar.month_name[month]
    print(f"\n## {month_name} {year} (Projected Revenue: {group['Projected_Revenue'].iloc[0]})")
    
    for _, event in group.iterrows():
        start_str = event['Start_Date'].strftime('%b %d (%a)')
        end_str = event['End_Date'].strftime('%b %d (%a)')
        print(f"  Sale #{event['Event']}: {start_str} to {end_str} ({event['Duration']} days)")

# Export the results to CSV
sale_planning_df.to_csv('optimal_sale_duration_plan.csv', index=False)
print("\nExported sale plan to 'optimal_sale_duration_plan.csv'")
```
